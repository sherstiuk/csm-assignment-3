A function to get all the needed information for analysis:

```python
def get_stats(df):
    #ht = [] 
    ht_flat = [] # all hashtags
    ht_pairs = [] # all hashtag pairs
    ht_1 = 0 # no. of tweets with at least one hashtag
    ht_2 = 0 # no. of tweets with more than one hashtag
    media = 0 # no. of tweets containing media
    url = 0 # no. of tweets containing links

    for index, row in df.iterrows():
        if 'urls' in row.entities:
            if len(row.entities['urls']) > 0:
                url += 1
        if 'media' in row.entities:
            if len(row.entities['media']) > 0:
                media += 1
        if 'hashtags' in row.entities:
            if len(row.entities['hashtags']) > 0:
                ht_1 += 1
                ht_this = []
                for h in row.entities['hashtags']:
                    ht_this.append(h['text'])
                    ht_flat.append(h['text'])
                if len(row.entities['hashtags']) > 1:
                    ht_2 += 1
                    for p in combinations(ht_this,2):
                        ht_pairs.append(p)
                #ht.append(ht_this)

    c = Counter(ht_flat)
    c2 = Counter(ht_pairs)
    
    top_50_ht = c.most_common(50) # 50 most frequent hashtags
    #top_20_ht = c.most_common(20)
    top_50_ht_pairs = c2.most_common(50) # 50 most frequent hashtag pairs
    
    return (ht_flat, ht_1, ht_2, media, url, top_50_ht, top_50_ht_pairs)
```

### Part 1

#### SAMPLE 1

```python
s1_ht, s1_ht_1, s1_ht_2, s1_w_media, s1_w_url, s1_top_50_ht, s1_top_50_ht_pairs = get_stats(sample_1)
```

```python
s1_no = len(sample_1)
print('* Total number of Tweets:\t {}'.format(s1_no))
print('* Total number and percentage of Tweets that contain at least 1 hashtag: \t{}, \t{}%'.format(s1_ht_1, (100*s1_ht_1/s1_no)))
print('* Total number and percentage of Tweets that contain 2 or more hashtags: \t{}, \t{}%'.format(s1_ht_2, (100*s1_ht_2/s1_no)))
print('* Total number and percentage of Tweets that contain media: \t{}, \t {}%'.format(s1_w_media, (100*s1_w_media/s1_no)))
print('* Total number and percentage of Tweets that contain URLs: \t{}, \t {}%'.format(s1_w_url, (100*s1_w_url/s1_no)))
```

```python
s1_50_ht_df = pd.DataFrame(s1_top_50_ht, columns = ['hashtag', 'frequency'])
s1_50_ht_df.index += 1
print('*  50 most frequent hashtags:')
s1_50_ht_df
```

```python
s1_50_p_df = pd.DataFrame(s1_top_50_ht_pairs, columns = ['hashtag pair', 'frequency'])
s1_50_p_df.index += 1
print('*  50 most frequent hashtag pairs:')
s1_50_p_df
```
